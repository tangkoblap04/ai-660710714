```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('student.csv')
```

```python
data.isnull().sum()
```

```python
data
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data.loc[data['GPA'] < 0, 'GPA'] = np.nan
data.loc[data['GPA'] > 4, 'GPA'] = 4.0
mean_gpa = data['GPA'].mean()
data['GPA'].fillna(mean_gpa, inplace=True)
```

```python
data
```

```python
faculty_mapping = {
    'วิทย์': 'วิทยาศาสตร์',
    'วิทยาศาสตร์': 'วิทยาศาสตร์',
    'ศิลปศาสตร์': 'ศิลปศาสตร์',
    'Arts': 'ศิลปศาสตร์',
    'วิศวะ': 'วิศวกรรมศาสตร์',
    'Liberal Arts' : 'ศิลปศาสตร์',
    'Sci' : 'วิทยาศาสตร์',
    'วิศวฯ' : 'วิศวกรรมศาสตร์',
    'คณะวิทยาศาสตร์' : 'วิทยาศาสตร์',
    'คณะศิลปศาสตร์' : 'ศิลปศาสตร์',
    'คณะวิศวกรรมศาสตร์' : 'วิศวกรรมศาสตร์',
    'ศิลป์' : 'ศิลปศาสตร์',
    'Engineering' : 'วิศวกรรมศาสตร์',
    'วิศวกรรม' : 'วิศวกรรมศาสตร์',
    'วิศวกรรมศาสตร์' : 'วิศวกรรมศาสตร์'
}
```

```python
data['คณะ'] = data['คณะ'].map(faculty_mapping)
```

```python
data
```

```python
data['เพศ'] = data['เพศ'].replace({
    'ช': 'ชาย',
    'M' : 'ชาย',
    'ญ': 'หญิง',
    'Female' : 'หญิง',
    'F' : 'หญิง'
})
```

```python
data
```

```python

data.loc[data['อายุ'] > 100, 'อายุ'] = np.nan
mean_age = data['อายุ'].mean()
data['อายุ'].fillna(mean_age, inplace=True)
data['อายุ'] = data['อายุ'].round().astype(int)
```
